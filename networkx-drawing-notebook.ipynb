## 1. Basic Drawing

All drawing in NetworkX can handled by the ``draw`` function. Every algorithm has its own convenience function but all of those functions are special cases of the ``draw`` function and here we will only consider that function. Different layouts can be acheived by calling specific layout functions directly and passing coordinates generated by these functions to the ``draw`` function.

```python
import networkx as nx
```

```python
G = nx.petersen_graph()
nx.draw(G)
```

The ``draw`` function is also highly configurable. Keyword parameters can be specified to customise things like node colours, edge widths and whether the graph is displayed with labels or not. Because it is quite likely that these configuration options are going to be used on several graphs it makes sense to use an alternative approach and put all of the option configurationsn into a dictionary and then pass the dictionary as a parameter to ``draw``.

```python
options = {
 'with_labels': False,
 'node_color': 'black',
 'node_size': 200,
 'width': 3,
}
```

```python
nx.draw(G, **options)
```

```python
pos_shell = nx.shell_layout(G, [range(5,10), range(5)])
nx.draw(G, pos_shell, **options)
```

## 2. Simple Layouts - graphs with few vertices

NetworkX has several layout algorithms.

- Circular
- Spectral
- Spring
- Shell
- Random
- Graphviz (through pygraphviz)

Here we will look at the first four of these algorithms. To give a quick impression of each of them consider the next figure which has four drawings of the Petersen Graph. One done by each of the first four layout algorithms in the above list.

The shell layout here is familiar. The ``shell_layout`` function takes an optional list argument. The elements of the list are lists of vertices. The vertices are then arranged in concentric shells according to the elements of these lists.

We configure the size of pylab figures to make them a little larger than the default.

```python
bigfigs()
```

```python
pos_circular = nx.circular_layout(G)
pos_spectral = nx.spectral_layout(G)
pos_spring = nx.spring_layout(G)
pos_shell = nx.shell_layout(G, [range(5,10), range(5)])

plt.subplot(221)
nx.draw(G, pos_circular, **options)

plt.subplot(222)
nx.draw(G, pos_spectral, **options)

plt.subplot(223)
nx.draw(G, pos_spring, **options)

plt.subplot(224)
nx.draw(G, pos_shell, **options)
```

The ``plt.suplot`` command will be familiar to anyone with experience of Matlab. It takes an integer argument whose first two digits are interpreted as number of rows and columns in a rectangular grid layout and whose third argument is then interpreted as the position for the current figure. Numbering goes down the grid from left to right.

## 2. Graphs with a few more vertices

The options we have chosen so far have been suitable for graphs with a few nodes. When we are drawing larger graphs it makes sense to choose smaller node sizes and thinner edges so that the graph doesn't emerge as a large black smudge.
