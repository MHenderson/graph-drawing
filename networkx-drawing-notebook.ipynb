## 1. Basic Layout

NetworkX has several layout algorithms.

- Circular
- Spectral
- Spring
- Shell
- Random
- Graphviz (through pygraphviz)

We don't look at the last two here.

```python
G = nx.chvatal_graph()

pos_circular = nx.circular_layout(G)
pos_spectral = nx.spectral_layout(G)
pos_spring = nx.spring_layout(G)
pos_shell = nx.shell_layout(G, [[0,2,3,4,8,9,10,6],[1,5,11,7]])

plt.subplot(221)
nx.draw(G, pos_circular, **options)

plt.subplot(222)
nx.draw(G, pos_spectral, **options)

plt.subplot(223)
nx.draw(G, pos_spring, **options)

plt.subplot(224)
nx.draw(G, pos_shell, **options)
```

## 2. Graphs with a few more vertices

```python
options_1 = {
           'with_labels':False,
           'node_color': 'black',
           'node_size': 20,
           'width':1,
           }
```

```python
barbell = nx.barbell_graph(10,10)

pos = nx.shell_layout(barbell)
nx.draw(barbell, pos, **options_1)
```

```python
lollipop = nx.lollipop_graph(10,20)
pos = nx.spring_layout(lollipop)
nx.draw(lollipop, pos, **options_1)
```

## 4. Importing layouts from Gephi

```python
options_3 = {
           'with_labels':False,
           'node_color': 'black',
           'node_size': 5,
           'width':1,
           'arrows': False
           }
```

```python
gephi_graph = nx.read_graphml("/home/matthew/tmp/red.graphml")

nx.draw(gephi_graph, **options_3)
```

```python
pos = dict([(v,(gephi_graph.node[v]['x'], gephi_graph.node[v]['y'])) for v in gephi_graph.nodes()])
nx.draw(gephi_graph, pos = pos, **options_3)
```
