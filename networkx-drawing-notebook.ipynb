## 1. Basic Drawing

All drawing in NetworkX can handled by the ``draw`` function. Every layout algorithm also has its own convenience function but all of those functions are special cases of the ``draw`` function and here we will only consider that function. Different layouts can be acheived by calling specific layout functions directly and passing coordinates generated by these functions to the ``draw`` function.

```python
import networkx as nx
```

NetworkX makes many named graphs available. For example, the Petersen Graph can be generated by calling the ``petersen_graph`` function.

```python
G = nx.petersen_graph()
```

To draw the Petersen Graph with a circular layout requires that we pass the ``draw`` function a parameter giving the coordinates of the vertices in a circular layout. These coordinates are the output of the ``circular_layout`` function which takes the graph as a parameter.

```python
setfigsize(5, 5)

nx.draw(G, nx.circular_layout(G))
```

The ``draw`` function can be configured to produce graph drawings with different visual properties. Keyword parameters can be specified to customise things like node colours, edge widths and whether the graph is displayed with labels or not.

```python
nx.draw(G, nx.circular_layout(G), with_labels=False, node_color='black')
```

Because it is quite likely that these configuration options are going to be used on several graphs it makes sense to use an alternative approach and put all of the option configurationsn into a dictionary and then pass the dictionary as a parameter to ``draw``.

```python
options = {
 'with_labels': False,
 'node_color': 'black',
 'node_size': 200,
 'width': 3,
}
```

```python
nx.draw(G, nx.circular_layout(G), **options)
```

The above drawing of the Petersen Graph is not easily recognised as a drawing of the Petersen Graph. One of the more recognisable drawings has one of the five cycles in the shape of a regular pentagon with another five cycle as a star in the interior and joined to the outer cycle by five spokes. We can achieve this sort of layout in NetworkX using the shell layout algorithm.

The ``shell_layout`` function takes an optional list argument. The elements of the list are lists of vertices. The vertices are then arranged in concentric shells according to the elements of these lists.

```python
pos = nx.shell_layout(G, [range(5,10), range(5)])
nx.draw(G, pos, **options)
```

## 2. Simple Layouts - graphs with few vertices

NetworkX has several layout algorithms.

- Circular
- Shell
- Spring
- Spectral
- Random
- Graphviz (through pygraphviz)

We have seen already examples of the circular and shell layouts. The next figure reproduces those two drawings from above along with examples of the spring and spectral layouts.

```python
setfigsize(10, 10)

plt.subplot(221)
nx.draw_circular(G, **options)

plt.subplot(222)
nx.draw_shell(G, nlist=[range(5,10), range(5)], **options)

plt.subplot(223)
nx.draw_spectral(G, **options)

plt.subplot(224)
nx.draw_spring(G, **options)
```

The ``plt.suplot`` command will be familiar to anyone with experience of Matlab. It takes an integer argument whose first two digits are interpreted as number of rows and columns in a rectangular grid layout and whose third argument is then interpreted as the position for the current figure. Numbering goes down the grid from left to right.

## 2. Graphs with a few more vertices

The options we have chosen so far have been suitable for graphs with a few nodes. When we are drawing larger graphs it makes sense to choose smaller node sizes and thinner edges so that the graph doesn't emerge as a large black smudge.

```python
options_1 = {
 'with_labels': False,
 'node_color': 'black',
 'node_size': 50,
 'width': 1,
}
```

```python
setfigsize(6, 6)

G = nx.dodecahedral_graph()
nx.draw_shell(G, nlist = [[2,3,4,5,6],[8,1,0,19,18,17,16,15,14,7],[9,10,11,12,13]], **options_1)
```

## 3. Graphs with many more vertices

```python
options_2 = {
 'with_labels': False,
 'node_color': 'grey',
 'node_size': 10,
 'linewidths': 0,
 'width': 0.1,
}
```

```python
setfigsize(18, 9)

G = nx.complete_graph(25)
plt.subplot(121)
nx.draw_circular(G, **options_2)

G = nx.barabasi_albert_graph(100, 3)
plt.subplot(122)
nx.draw_circular(G, **options_2)
```
