# Graph Drawing in NetworkX

```python
import random
random.seed(0)

pylab.rcParams['figure.figsize'] = 7, 7
```

```python
import networkx as nx
```

```python
K_3_5 = nx.complete_bipartite_graph(3,5)

options = {
           'with_labels':False,
           'node_color': 'black',
           'node_size': 250,
           'width':3,
           }
pos = nx.spectral_layout(K_3_5)

nx.draw(K_3_5, pos, **options)
```

```python




pos = nx.circular_layout(K_3_5)
nx.draw(K_3_5, pos = pos, **options)
```

```python
barbell = nx.barbell_graph(10,10)

options_1 = {
           'with_labels':False,
           'node_color': 'black',
           'node_size': 20,
           'width':1,
           }

pos = nx.shell_layout(barbell)
nx.draw(barbell, pos, **options_1)
```

```python
lollipop = nx.lollipop_graph(10,20)
pos = nx.spring_layout(lollipop)
nx.draw(lollipop, pos, **options_1)
```

```python
er = nx.erdos_renyi_graph(100,0.15)
nx.draw(er, **options_1)
```

```python
ws = nx.watts_strogatz_graph(30,3,0.1)
nx.draw(ws, **options_1)
```

```python
ba = nx.barabasi_albert_graph(100,5)
nx.draw(ba, **options_1)
```

```python
red = nx.random_lobster(100,0.9,0.9)
nx.draw(red, **options_1)
```

```python
options_2 = {
           'with_labels':False,
           'node_color': 'black',
           'node_size': 20,
           'width':1,
           'arrows': False
           }

gephi_graph = nx.read_graphml("/home/matthew/tmp/red.graphml")

nx.draw(gephi_graph, **options_2)
```

```python
pos = dict([(v,(gephi_graph.node[v]['x'], gephi_graph.node[v]['y'])) for v in gephi_graph.nodes()])
nx.draw(gephi_graph, pos = pos, **options_2)

```
